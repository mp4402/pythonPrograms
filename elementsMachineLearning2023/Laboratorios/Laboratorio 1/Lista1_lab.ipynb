#### Datos Faltantes

```python
df.head()
```

```python
df.isna().sum()
```

```python
from sklearn.impute import SimpleImputer
```

```python
imp_mean = SimpleImputer(missing_values=np.nan, strategy='mean')
imp_mode = SimpleImputer(missing_values=np.nan, strategy='most_frequent')
```

```python
df['bill_length_mm'] = imp_mean.fit_transform(df[['bill_length_mm']])
```

```python
df['bill_depth_mm'] = imp_mean.fit_transform(df[['bill_depth_mm']])
```

```python
df['flipper_length_mm'] = imp_mean.fit_transform(df[['flipper_length_mm']])
```

```python
df['body_mass_g'] = imp_mean.fit_transform(df[['body_mass_g']])
```

```python
df['sex'] = imp_mode.fit_transform(df[['sex']])
```

## Heatmap (Join Density)

```python
plt.figure()
g = sns.JointGrid(data=df, x="body_mass_g", y="bill_depth_mm", space=0)
g.plot_joint(sns.kdeplot, fill=True, clip=((2200, 6800), (10, 25)),
             thresh=0, levels=50, cmap="rocket")
g.plot_marginals(sns.histplot, color="#A3051A", alpha=1, bins=25)
```
