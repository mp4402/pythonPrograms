```python
#Carga de librerías
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sb


#Evitar alertas
from warnings import filterwarnings
filterwarnings('ignore')
```

```python
#Carga de archivos
company = ["Amazon","Apple","Facebook","Google","Microsoft","Netflix","Tesla","Uber","Walmart","Zoom"]
sp = pd.DataFrame([])
for i in range(0,len(company)):
    path = 'https://raw.githubusercontent.com/LuzAr123/Data_Science/main/ds/'+company[i]+'.csv'
    tmp = pd.read_csv(path, sep=',', header=0)
    tmp["Company"] = company[i]
    sp=pd.concat([sp,tmp],axis=0)
```
