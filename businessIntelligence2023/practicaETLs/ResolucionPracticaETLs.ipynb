```python
import pandas as pd
import json

playlist = json.load(open('mpd.slice.1000-1999.json'))


df = pd.json_normalize(playlist['playlists'], 'tracks', 
                  [
                      'name', 
                      'collaborative',
                      'pid',
                      'modified_at',
                      'num_tracks',
                      'num_albums',
                      'num_followers',
                      'num_edits',
                      'duration_ms',
                      'num_artists'
                  ],
                  meta_prefix='playlist_'
)
df
```

```python
df_playlist = (df.loc[:,['playlist_pid', 
          'playlist_name', 
          'playlist_collaborative', 
          'playlist_modified_at',
         'playlist_num_followers',
         'playlist_num_edits']]
 .drop_duplicates()
 .reset_index()
 .rename(columns={
     'index': 'sk_id',
     'playlist_pid': 'bk_id', 
          'playlist_name': 'name', 
          'playlist_collaborative': 'collaborative', 
          'playlist_modified_at': 'last_modified_at',
         'playlist_num_followers': 'num_followers',
         'playlist_num_edits': 'num_edits'
 })
 .assign(sk_id = lambda d: d.index+1,
        last_modified_at = lambda df: pd.to_datetime(df.last_modified_at, unit='s'))
)
df_playlist
```

```python
from sqlalchemy import create_engine

target = create_engine('postgresql+psycopg2://test:test123@localhost:5432/spotify_test')
```

```python
df_playlist.to_sql('d_playlists', con=target, index=False, if_exists='append')
```

```python
df_tracks = (df.loc[:,['track_uri', 
          'track_name', 
          'duration_ms']]
 .drop_duplicates()
 .reset_index()
 .rename(columns={
     'index': 'sk_id',
     'track_uri': 'bk_id', 
          'track_name': 'name', 
          'duration_ms': 'duration'
 })
 .assign(sk_id = lambda d: d.index+1)
)
df_tracks
```

```python
df_tracks.to_sql('d_tracks', con=target, index=False, if_exists='append')
```

```python
df_artist = (df.loc[:,['artist_uri', 
          'artist_name']]
 .drop_duplicates()
 .reset_index()
 .rename(columns={
     'index': 'sk_id',
     'artist_uri': 'bk_id', 
          'artist_name': 'name'
 })
 .assign(sk_id = lambda d: d.index+1)
)
df_artist
```

```python
df_artist.to_sql('d_artists', con=target, index=False, if_exists='append')
```

```python
df_album = (df.loc[:,['album_uri', 
          'album_name']]
 .drop_duplicates()
 .reset_index()
 .rename(columns={
     'index': 'sk_id',
     'album_uri': 'bk_id', 
          'album_name': 'name'
 })
 .assign(sk_id = lambda d: d.index+1)
)
df_album
```

```python
df_album.to_sql('d_albums', con=target, index=False, if_exists='append')
```

```python
df_f_playlist = (df
 .merge(df_playlist.loc[:,['sk_id', 'bk_id']], 
       left_on='playlist_pid', 
       right_on='bk_id', 
       how='inner',
       suffixes=('','_d_playlist'))
 .merge(df_tracks.loc[:,['sk_id', 'bk_id']],
       left_on='track_uri', 
       right_on='bk_id', 
       how='inner',
       suffixes=('_d_playlists','_d_tracks'))
 .merge(df_artist.loc[:,['sk_id', 'bk_id']],
       left_on='artist_uri', 
       right_on='bk_id', 
       how='inner',
       suffixes=('','_d_artist'))
  .merge(df_album.loc[:,['sk_id', 'bk_id']],
       left_on='album_uri', 
       right_on='bk_id', 
       how='inner',
       suffixes=('_d_artist','_d_album'))
 .loc[:,['sk_id_d_playlists',
        'sk_id_d_artist',
        'sk_id_d_album',
        'sk_id_d_tracks',
        'pos']]
 .rename(columns={
     "sk_id_d_playlists": "sk_playlist",
     "sk_id_d_artist": "sk_artist",
     "sk_id_d_album": "sk_album",
     "sk_id_d_tracks": "sk_track",
     "pos": "track_position"
 })
 .assign(id = lambda df: df.index+1)
)
df_f_playlist
    
```

```python
df_f_playlist.to_sql('f_playlist_tracks', con=target, index=False, if_exists='append')
```
